# 2. Директории

# З Загрузка данных

```python
# data
df_name = 'data.pqt'
df = pd.read_parquet(f'{proc_data_dir}{df_name}')

print(f'data size: {df.shape}')

# valid features
val_features_name = 'valid_features.xlsx'
val_features = pd.read_excel(f'{docs_dir}{val_features_name}')

val_features_ = val_features[val_features['valid_flag'] == 1]
print(f'valid_features size: {val_features.shape}')

# final_cols
data_cols = [*val_features_['attribute'], 'target']
print(f'final_features size: {len(data_cols)}')
```

# 4 Feature Engineering

```python
df_new = df.copy()

# преобразование категориальных переменных
df_new = preparing_cat_features(df_new)
print('cat_features: ok')

# добавление признаков по картам и суммам транзакций - с временными периодами (TransactionDT + TransactionAmt)
df_new = make_time_window_features(df_new)
print('time_features: ok')

# добавление признаков по картам и суммам транзакций + комбинации показателей
df_new = make_card_features(df_new)
print('card_features: ok')
```
